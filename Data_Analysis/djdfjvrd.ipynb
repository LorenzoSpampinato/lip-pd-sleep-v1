```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import normaltest, kruskal, f_oneway, mannwhitneyu
from statsmodels.stats.multicomp import pairwise_tukeyhsd


def create_mannwhitney_matrix(df, feature, phase):
    """
    Crea una matrice 4x4 con i p-value del test di Mann-Whitney U tra i gruppi durante una fase specifica.
    I valori sono in notazione scientifica.
    """
    groups = ['CTL', 'DNV', 'ADV', 'DYS']
    mw_matrix = pd.DataFrame(np.ones((4, 4)), index=groups, columns=groups)  # Matrice inizializzata con 1
    mw_matrix_text = pd.DataFrame(np.empty((4, 4), dtype=object), index=groups, columns=groups)  # Matrice per le stringhe

    for i in range(len(groups)):
        for j in range(i + 1, len(groups)):  # Solo triangolo superiore
            g1, g2 = groups[i], groups[j]
            g1_data = df[(df['Group'] == g1) & (df['Phase_Assigned'] == phase)][feature].dropna()
            g2_data = df[(df['Group'] == g2) & (df['Phase_Assigned'] == phase)][feature].dropna()

            if len(g1_data) > 0 and len(g2_data) > 0:
                _, p_value = mannwhitneyu(g1_data, g2_data)
                formatted_p_value = f"{p_value:.2e}"  # Notazione scientifica
                
                mw_matrix.loc[g1, g2] = p_value
                mw_matrix.loc[g2, g1] = p_value  # Specchia la matrice
                mw_matrix_text.loc[g1, g2] = formatted_p_value
                mw_matrix_text.loc[g2, g1] = formatted_p_value  # Specchia la matrice

    np.fill_diagonal(mw_matrix.values, 1)  # Diagonale a 1
    np.fill_diagonal(mw_matrix_text.values, '1.00e+00')  # Diagonale a '1.00e+00' in notazione scientifica
    return mw_matrix, mw_matrix_text

def plot_mannwhitney_heatmap(matrix, text_matrix, phase, feature):
    """
    Visualizza la matrice di Mann-Whitney U con una heatmap.
    """
    plt.figure(figsize=(8, 6))
    sns.heatmap(matrix, annot=text_matrix.values, fmt='', cmap='coolwarm', cbar=True)
    plt.title(f'Mann-Whitney U Matrix ({phase} Phase - {feature})')
    plt.show()

def analyze_feature(df, feature):
    results = {}

    for channel in df['Channel'].unique():
        print(f"\n //////////////////Processing Channel: {channel}//////////////////")
        channel_data = df[df['Channel'] == channel]

        for phase in ['Early', 'Late']:
            print(f"\nProcessing Phase: {phase}")
            phase_data = channel_data[channel_data['Phase_Assigned'] == phase]
            groups = phase_data['Group'].unique()
            data = [phase_data[phase_data['Group'] == group][feature].dropna() for group in groups]
            
            # Creazione della tabella descrittiva (colonne: statistiche, righe: gruppi)
            descriptive_table = phase_data.groupby('Group')[feature].describe().drop(columns=['min', 'max']).T
            print(f"\nDescriptive Statistics for {feature} ({phase} Phase):")
            print(descriptive_table)
            print()

            # Test di normalit
            normality = {group: normaltest(group_data)[1] for group, group_data in zip(groups, data)}
            print(f"Normality Test (p-value) for {phase} Phase:")
            for group, p_value in normality.items():
                print(f"  {group}: p-value = {p_value:.4e}")
            print()

            # Scelta del test statistico (ANOVA o Kruskal-Wallis)
            if all(p > 0.05 for p in normality.values()):  # Se tutte le distribuzioni sono normali
                stat, p = f_oneway(*data)
                test_type = "ANOVA"
            else:
                stat, p = kruskal(*data)
                test_type = "Kruskal-Wallis"

            print(f"Test Results for {test_type}:")
            print(f"  Test Statistic = {stat:.4f}, p-value = {p:.4e}")
            print()

            post_hoc_results = None
            if p < 0.05:
                combined_data = phase_data[[feature, 'Group']].dropna()

                if test_type == "ANOVA":
                    post_hoc = pairwise_tukeyhsd(combined_data[feature], combined_data['Group'])
                    post_hoc_results = post_hoc.summary().data[1:]
                    post_hoc_results = pd.DataFrame(post_hoc_results, columns=['Group1', 'Group2', 'P-Value', 'Lower', 'Upper', 'Reject'])
                else:
                    # Mann-Whitney U test post-hoc
                    comparisons = []
                    p_values = []

                    for i in range(len(groups)):
                        for j in range(i + 1, len(groups)):
                            g1 = groups[i]
                            g2 = groups[j]
                            g1_data = phase_data[phase_data['Group'] == g1][feature].dropna()
                            g2_data = phase_data[phase_data['Group'] == g2][feature].dropna()

                            if len(g1_data) > 0 and len(g2_data) > 0:
                                _, p_value = mannwhitneyu(g1_data, g2_data)
                                formatted_p_value = f"{p_value:.2e}"
                                comparisons.append((g1, g2, formatted_p_value))

                    post_hoc_results = pd.DataFrame(comparisons, columns=['Group1', 'Group2', 'P-Value'])

                print(f"Post-hoc Results for {feature} ({test_type}):\n", post_hoc_results)

            # Salva i risultati (senza min/max nelle descrizioni)
            results[(channel, phase)] = {
                'Descriptive': phase_data.groupby('Group')[feature].describe().drop(columns=['min', 'max']),
                'Normality': normality,
                'Test': (test_type, stat, p),
                'PostHoc': post_hoc_results
            }


        # Creazione della matrice Mann-Whitney per Early e Late
        mw_matrix_early, mw_matrix_text_early = create_mannwhitney_matrix(channel_data, feature, 'Early')
        plot_mannwhitney_heatmap(mw_matrix_early, mw_matrix_text_early, 'Early', feature)

        mw_matrix_late, mw_matrix_text_late = create_mannwhitney_matrix(channel_data, feature, 'Late')
        plot_mannwhitney_heatmap(mw_matrix_late, mw_matrix_text_late, 'Late', feature)

        # Confronto tra Early e Late per ogni gruppo
        print("\nComparing Early vs Late for each group:")
        for group in groups:
            early_data = channel_data[(channel_data['Group'] == group) & (channel_data['Phase_Assigned'] == 'Early')][feature].dropna()
            late_data = channel_data[(channel_data['Group'] == group) & (channel_data['Phase_Assigned'] == 'Late')][feature].dropna()

            if len(early_data) > 0 and len(late_data) > 0:
                _, p_value = mannwhitneyu(early_data, late_data)
                print(f"Group: {group}, p-value (Early vs Late): {p_value:.2e}")

    return results


# Carica i dati
data = pd.read_csv(file_path)
results = analyze_feature(data, feature)
```

```python
def plot_early_and_late_phases_grouped_ordered(data, feature_name, channel_name,
                                               group_order=['CTL', 'DNV', 'ADV', 'DYS']):
    """
    Crea e mostra i boxplot per le fasi "Early" e "Late", raggruppando i pazienti per gruppo, per un singolo canale.
    Ogni grafico rappresenta un canale specifico, con boxplot per le fasi "Early" e "Late", e barplot per la media dei gruppi
    con intervallo SEM. Ora include anche la media di ciascun paziente.

    Parametri:
    - data: DataFrame con i dati.
    - feature_name: Nome della feature da plottare (colonna y).
    - channel_name: Canale da plottare.
    - group_order: Lista opzionale per ordinare i gruppi sull'asse x.
    """

    # Assicurati che la colonna 'Group' sia categorica con l'ordine desiderato
    data['Group'] = pd.Categorical(data['Group'], categories=group_order, ordered=True)

    # Filtra i dati per il canale specifico
    channel_data = data[data['Channel'] == channel_name]

    # Filtra i dati per le fasi "Early" e "Late"
    combined_data = channel_data[channel_data['Phase_Assigned'].isin(["Early", "Late"])]

    # Conta il numero di epoche Early e Late per ogni soggetto
    epoch_counts = combined_data.groupby(['Subject', 'Phase_Assigned']).size().unstack(fill_value=0)

    # Stampa il numero di epoche per ogni soggetto
    print("\nNumero di epoche per ciascun soggetto:")
    print(epoch_counts)

    # Trova gli indici delle epoche Early e Late per ogni soggetto
    early_indices = combined_data[combined_data['Phase_Assigned'] == "Early"].groupby('Subject').apply(
        lambda x: list(x.index))
    late_indices = combined_data[combined_data['Phase_Assigned'] == "Late"].groupby('Subject').apply(
        lambda x: list(x.index))

    # Stampa gli indici delle epoche
    print("\nIndici delle epoche Early per ciascun soggetto:")
    print(early_indices)
    print("\nIndici delle epoche Late per ciascun soggetto:")
    print(late_indices)

    # Crea una mappa di colori unica per ogni paziente
    unique_subjects = sorted(combined_data['Subject'].unique())  # Ordina i pazienti
    subject_colors = sns.color_palette("tab20", n_colors=len(unique_subjects))
    color_map = {subject: color for subject, color in zip(unique_subjects, subject_colors)}

    # Creazione del grafico
    plt.figure(figsize=(16, 8))
    x_offset = 0.4
    tick_positions = []
    tick_labels = []

    # Traccia i dati per ogni gruppo
    for group_idx, group in enumerate(combined_data['Group'].cat.categories):
        group_data = combined_data[combined_data['Group'] == group]
        group_data = group_data.sort_values(by=["Subject"])

        early_data = group_data[group_data['Phase_Assigned'] == "Early"]
        late_data = group_data[group_data['Phase_Assigned'] == "Late"]

        early_means = early_data.groupby('Subject')[feature_name].mean()
        late_means = late_data.groupby('Subject')[feature_name].mean()

        early_group_mean = early_means.mean()
        late_group_mean = late_means.mean()
        early_sem = early_means.sem()
        late_sem = late_means.sem()

        sns.boxplot(
            x=np.full(len(early_data), group_idx * 2 - x_offset),
            y=early_data[feature_name],
            hue=early_data['Subject'],
            data=early_data,
            palette=color_map,
            dodge=True,
            width=0.4,
            boxprops=dict(alpha=0.8),
            showfliers=True
        )

        sns.boxplot(
            x=np.full(len(late_data), group_idx * 2 + x_offset),
            y=late_data[feature_name],
            hue=late_data['Subject'],
            data=late_data,
            palette=color_map,
            dodge=True,
            width=0.4,
            boxprops=dict(alpha=0.8),
            showfliers=True
        )

        plt.bar(
            group_idx * 2 - x_offset, early_group_mean, color='red', width=0.2, zorder=3
        )
        plt.bar(
            group_idx * 2 + x_offset, late_group_mean, color='blue', width=0.2, zorder=3
        )

        plt.errorbar(
            group_idx * 2 - x_offset, early_group_mean, yerr=early_sem, fmt='none', color='black', capsize=5, zorder=4
        )
        plt.errorbar(
            group_idx * 2 + x_offset, late_group_mean, yerr=late_sem, fmt='none', color='black', capsize=5, zorder=4
        )

        plt.scatter(
            np.full(len(early_means), group_idx * 2 - x_offset),
            early_means.values,
            color=[color_map[subj] for subj in early_means.index],
            edgecolor='black',
            linewidth=0.5,
            zorder=5,
            s=50
        )

        plt.scatter(
            np.full(len(late_means), group_idx * 2 + x_offset),
            late_means.values,
            color=[color_map[subj] for subj in late_means.index],
            edgecolor='black',
            linewidth=0.5,
            zorder=5,
            s=50
        )

        tick_positions.append(group_idx * 2)
        tick_labels.append(group)

    plt.xticks(ticks=tick_positions, labels=tick_labels, fontsize=12)
    plt.xlabel("Gruppi", fontsize=14)
    plt.ylabel(feature_name, fontsize=14)
    plt.title(f"Boxplot Early e Late per gruppo ({feature_name}) - Canale: {channel_name}", fontsize=16)

    # Rimuovere la legenda duplicata e posizionarla fuori dal grafico
    handles, labels = plt.gca().get_legend_handles_labels()
    unique_legend = dict(zip(labels, handles))  # Rimuove i duplicati
    plt.legend(unique_legend.values(), unique_legend.keys(), bbox_to_anchor=(1.05, 1), loc='upper left', title="Pazienti")

    plt.tight_layout()
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()



# Carica i dati
file_path = r"C:\Users\Lorenzo\Desktop\prova statistica\5 canali\_no_mean_features_specific_channels_5_aggregated_with_phases.csv"
data = pd.read_csv(file_path)





'''
# Ciclo per analizzare ogni caratteristica
for feature in feature_to_analyze:
    print(f"\n### Analisi per caratteristica: {feature} ###")
    results = analyze_feature(data, feature)
'''
# Specifica i canali selezionati per i grafici (uno o pi)
selected_channels = ['E224']  # Aggiungi tutti i canali di interesse

# Genera i grafici per ogni canale e caratteristica
for channel_name in selected_channels:
    for feature_name in feature:
        print(f"\n### Grafico per caratteristica: {feature_name}, Canale: {channel_name} ###")
        # Grafico 2: Subplot organizzati per pazienti e gruppo
        plot_early_and_late_phases_grouped_ordered(data.copy(),feature_name=feature_name,channel_name=channel_name,group_order=['CTL', 'DNV', 'ADV', 'DYS'])

```

```python
from sklearn.preprocessing import MinMaxScaler
def plot_early_and_late_barplots_minmaxscaler(data, feature_name, channel_name, group_order=['CTL', 'DNV', 'ADV', 'DYS']):
    """
    Creates and displays barplots for the "Early" and "Late" phases, grouping patients by group.
    The barplots represent the group means, and colored dots represent the individual patient means.

    Parameters:
    - data: DataFrame with the data.
    - feature_name: Name of the feature to plot (y column).
    - channel_name: Channel to plot.
    - group_order: Optional list to order the groups on the x-axis.
    """
    # **Apply MinMaxScaler to the entire dataset**
    scaler = MinMaxScaler()
    data[feature_name] = scaler.fit_transform(data[[feature_name]])

    # Ensure that the 'Group' column is categorical with the desired order
    data['Group'] = pd.Categorical(data['Group'], categories=group_order, ordered=True)

    # Filter data by the specified channel
    channel_data = data[data['Channel'] == channel_name]

    # Filter data by "Early" and "Late" phases
    combined_data = channel_data[channel_data['Phase_Assigned'].isin(["Early", "Late"])]

    # Plot creation
    plt.figure(figsize=(12, 6))
    x_offset = 0.3  # Distance between Early and Late bars
    tick_positions = []
    tick_labels = []

    # Create a unique color map for each patient
    unique_subjects = sorted(combined_data['Subject'].unique())  # Sort the patients
    subject_colors = sns.color_palette("Paired", n_colors=len(unique_subjects))
    color_map = {subject: color for subject, color in zip(unique_subjects, subject_colors)}

    # Variable to handle patient legends (to avoid duplicates)
    patient_legend_handles = []

    # Plot data for each group
    for group_idx, group in enumerate(combined_data['Group'].cat.categories):
        group_data = combined_data[combined_data['Group'] == group]

        early_data = group_data[group_data['Phase_Assigned'] == "Early"]
        late_data = group_data[group_data['Phase_Assigned'] == "Late"]

        early_means = early_data.groupby('Subject')[feature_name].mean()
        late_means = late_data.groupby('Subject')[feature_name].mean()

        early_group_mean = early_means.mean()
        late_group_mean = late_means.mean()

        # **Barplot for group means**
        plt.bar(
            group_idx * 2 - x_offset, early_group_mean, color='red', width=0.5, label="Early" if group_idx == 0 else "",
            zorder=3
        )
        plt.bar(
            group_idx * 2 + x_offset, late_group_mean, color='blue', width=0.5, label="Late" if group_idx == 0 else "",
            zorder=3
        )

        # Add dots for each patient with a unique color
        for idx, subject in enumerate(early_means.index):
            plt.scatter(
                [group_idx * 2 - x_offset], [early_means.loc[subject]], color=color_map[subject], zorder=4, alpha=1,
                edgecolor='black', linewidth=1.5, s=70
            )

        for idx, subject in enumerate(late_means.index):
            plt.scatter(
                [group_idx * 2 + x_offset], [late_means.loc[subject]], color=color_map[subject], zorder=4, alpha=1,
                edgecolor='black', linewidth=1.5, s=70
            )

        tick_positions.append(group_idx * 2)
        tick_labels.append(group)

    # Customizing the X-axis
    plt.xticks(ticks=tick_positions, labels=tick_labels, fontsize=12)
    plt.xlabel("Groups", fontsize=14)
    plt.ylabel(feature_name, fontsize=14)
    plt.title(f"Barplot Early and Late per Group ({feature_name}) - Channel: {channel_name} - Scaler: MinMaxScaler", fontsize=16)

    # Create a unique legend for the patients (to avoid duplicates)
    for subject, color in color_map.items():
        patient_legend_handles.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=subject))

    # Add the Early/Late legend
    plt.legend(handles=patient_legend_handles + [plt.Line2D([0], [0], color='red', lw=4, label='Early'),
                                                 plt.Line2D([0], [0], color='blue', lw=4, label='Late')],
               bbox_to_anchor=(1.05, 1), loc='upper left', title="Patients")

    plt.tight_layout()
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()
# Specifica i canali selezionati per i grafici (uno o pi)
selected_channels = ['E224']  # Aggiungi tutti i canali di interesse

# Genera i grafici per ogni canale e caratteristica
for channel_name in selected_channels:
    for feature_name in feature:
        print(f"\n### Grafico per caratteristica: {feature_name}, Canale: {channel_name} ###")
    
        plot_early_and_late_barplots_minmaxscaler(data.copy(), feature_name=feature_name, channel_name=channel_name,group_order=['CTL', 'DNV', 'ADV', 'DYS'])

```
